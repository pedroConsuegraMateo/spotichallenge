```python
def get_artists_id(base_url, resource, main_id = '/0oSGxfWSnnOXhD2fKuz2Gy'):
    url = base_url + resource + main_id + '/related-artists'
    response_related = requests.get(url, headers=headers)
    results_related = response_related.json()
    data_result_related = pd.DataFrame(results_related['artists'])
    
    return data_result_related['id']
```

```python
def get_top_tracks(base_url, resource, artist_id):
    url_artist = base_url + resource + '/' + artist_id + '/top-tracks' + '?market=ES'
    response = requests.get(url_artist, headers=headers)
    results = response.json()
    data_result = pd.DataFrame(results['tracks'])

    return data_result
    
```

```python
def funcion(x):
    return x[0]['name']
```

```python
artists = get_artists_id(base_url, resource)
artists
a = []
for i in artists:
    a.append(i)
print(a)

top_tracks = []

for i in a:
    top_tracks.append(get_top_tracks(base_url, resource, i)) 

#print(top_tracks[0]['artists'][0][0]['name'])
top_tracks2 = pd.concat(top_tracks)
top_tracks2['artists'] = top_tracks2['artists'].apply(funcion)
top_tracks2[['artists', 'name']]
```
